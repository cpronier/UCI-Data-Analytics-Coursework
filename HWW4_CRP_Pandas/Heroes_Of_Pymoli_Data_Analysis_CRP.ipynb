```python
#Import libraries needed for analysis
import pandas as pd
import numpy as np

#Load data files for analysis
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
pymoli_data = pd.read_csv(file_to_load)
```

```python
#View the first few rows to get a sense of the dataframe and list column names needed.
pymoli_data.head()
```

```python
#Determine the number of players by counting the unique screen names listed in the dataframe
total_players=pymoli_data['SN'].nunique()

#Create dataframe for output to display
total_players_summary=pd.DataFrame([{'Total Players':total_players}])
total_players_summary
```

```python
#The number is different based on using either Item Name or Item ID to determine unique number of items.
unique_items=pymoli_data['Item ID'].nunique()

#New variable for average price using mean function
average_price=pymoli_data['Price'].mean()

#Currency formatting for new average price variable
average_price='${:,.2f}'.format(average_price)

#New variable to count number of purchases based on the purchase ID.
number_purchases=pymoli_data['Purchase ID'].count()

#New variable for sum of all sales
total_revenue=pymoli_data['Price'].sum()

#Currency formatting for the revenue data
total_revenue='${:,.2f}'.format(total_revenue)

#Create dataframe with above variables
summary_df=pd.DataFrame([{'Unique Items':unique_items, 'Average Price':average_price, 'Number of Purchases':number_purchases, 'Total Revenue': total_revenue}])
summary_df
```
